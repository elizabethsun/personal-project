```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import BernoulliNB
pd.set_option('display.max_colwidth', -1) 
```

```python
yelp_raw = pd.read_csv('/Users/elizabeth/Desktop/Thinkful Data Science/Datasets/sentiment labelled sentences/yelp_labelled.txt', delimiter= '\t', header=None)
yelp_raw.columns = ['review', 'sentiment']

keywords = ['not good', 'nasty', 'angry', 'never', 'disappointing', 'poor', 'not', 'disgusted', 'slow', 
            'bad' ,'underwhelming' ,'suck', 'flop', 'avoid', 'horrible', 'wasted','no']

for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    yelp_raw[str(key)] = yelp_raw.review.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )


data = yelp_raw[keywords]
target = yelp_raw['sentiment']

bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)
```

```python
yelp_raw.loc[yelp_raw['sentiment']==1].count()[0]/yelp_raw.shape[0]
```

```python
yelp_raw.loc[yelp_raw['sentiment']==0].iloc[300:350, 0:2]
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(target, y_pred)
```

```python
from sklearn.model_selection import train_test_split
# Use train_test_split to create the necessary training and test groups
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=20)
print('With 20% Holdout: ' + str(bnb.fit(X_train, y_train).score(X_test, y_test)))
print('Testing on Sample: ' + str(bnb.fit(data, target).score(data, target)))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(bnb, data, target, cv=10)
```
